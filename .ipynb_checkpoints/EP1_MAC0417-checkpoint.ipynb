```python
caminho_imagem = 'dados/001.jpg'

# Verificar se o arquivo existe
if os.path.exists(caminho_imagem):
    # Carregar a imagem
    with open(caminho_imagem, 'rb') as f:
        imagem = plt.imread(f)

    # Mostrar a imagem
    plt.imshow(imagem)
    plt.axis('off')  # Opcional: desativar os eixos
    plt.show()
else:
    print("O arquivo de imagem não existe.")
```
